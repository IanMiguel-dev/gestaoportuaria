```python
import matplotlib.pyplot as plt
import numpy as np

# Simulando dados de magnitude do campo magnético (em nanoTesla)
dias = np.arange(0, 100)
magnitude_magnetica = 50000 + 300 * np.random.randn(100)  # média ~50000 nT

# Criar gráfico
plt.plot(dias, magnitude_magnetica)
plt.fill_between(dias, magnitude_magnetica, color='green', alpha=0.5)
plt.title("Magnitude do Campo Magnético da Terra")
plt.xlabel("Dias")
plt.ylabel("nT (nanoTesla)")
plt.grid(True)
plt.show()
```

```python
# prompt: crie um grafico de abalo sismiticos

import matplotlib.pyplot as plt
import numpy as np

# Simulando dados de abalos sísmicos (magnitude)
dias = np.arange(0, 100)
magnitude_terremotos = 2 + 1 * np.random.randn(100)  # Magnitude média ~2

# Criar gráfico
plt.plot(dias, magnitude_terremotos)
plt.fill_between(dias, magnitude_terremotos, color='red', alpha=0.3) #alpha controla a transparencia
plt.title("Magnitude de Abalos Sísmicos")
plt.xlabel("Dias")
plt.ylabel("Magnitude")
plt.grid(True)
plt.show()
```
